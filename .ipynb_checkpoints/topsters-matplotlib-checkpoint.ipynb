```python
# Topsters Practice

import os
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

filepath = os.path.join("Data","topsters.csv")

topsters_DF = pd.read_csv(filepath)
topsters_DF.sample(15)
```

```python
print(f"{len(topsters_DF['COUNTRY'].unique())} countries represented:")
topsters_DF['COUNTRY'].value_counts()
```

```python
albumcount = len(topsters_DF['TITLE'].unique())
artistcount = len(topsters_DF['ARTIST'].unique())
genrecount = len(topsters_DF['GENRE'].unique())
yearspan = topsters_DF['YEAR'].max() - topsters_DF['YEAR'].min()
nationals = len(topsters_DF['COUNTRY'].unique())
summary = {"Total Number of Albums":albumcount,
                          "Total Number of Artists":artistcount,
                          "Total Genres Represented":genrecount,
                          "Range of Years" : yearspan,
                          "Number of Countries" : nationals}

summary_DF = pd.DataFrame([summary])
summary_DF
```

```python
# Locate and retrieve subset of Topsters for albums that came out in 2000 and beyond.
topsters00_DF = topsters_DF.loc[topsters_DF["YEAR"] > 1999]
total_2000s = len(topsters_DF.loc[topsters_DF["YEAR"] > 1999])
print(total_2000s)
topsters00_DF
```

```python
albumcount00 = topsters00_DF['TITLE'].count()
artistcount00 = len(topsters00_DF['ARTIST'].unique())
genrecount00 = len(topsters00_DF['GENRE'].unique())

summary00 = {"Total Number of Albums":albumcount00,
                          "Total Number of Artists":artistcount00,
                          "Total Genres Represented":genrecount00}

summary00_DF = pd.DataFrame([summary00])
print("Summary stats for albums released in 2000s")
summary00_DF
```

```python
# Using .groupby to show how many albums from each genre, from each country, we have:
genrecountry = pd.DataFrame(topsters_DF.groupby(["COUNTRY","GENRE"])["TITLE"].count())
genrecountry
```

```python
# Create a bar graph showing the distribution of albums by year
x_axis = np.arange(len(topsters_DF["YEAR"].unique()))
tick_loc = [value+0.4 for value in x_axis]
# df["column"].unique() returns an array.
# So we append .tolist() in order to make the array a list.
# Then we use sorted() at the beginning so we can have the list sorted in order.
years = sorted(topsters_DF["YEAR"].unique().tolist())
yearcount = pd.DataFrame(topsters_DF["YEAR"].value_counts().sort_index())

plt.figure(figsize=(25,15))
plt.xticks(tick_loc,years,rotation="vertical",fontsize=12)
plt.xlabel("Release Year",fontsize=20)
plt.ylabel("Number of Albums",fontsize=20)
plt.title("Topsters Breakdown: Albums per Year",fontsize=20)

plt.bar(x_axis,yearcount["YEAR"])

plt.savefig("Visuals/topsters_yearcount.jpg")
plt.show()
```

```python
# Create bar graph showing distribution of albums over genres
x_axis=x_axis = np.arange(len(topsters_DF["GENRE"].unique()))
tick_loc = [value for value in x_axis]
genres = sorted(topsters_DF["GENRE"].unique().tolist())
genrecount = pd.DataFrame(topsters_DF["GENRE"].value_counts().sort_index())

plt.figure(figsize=(20,10))
plt.xticks(tick_loc,genres,rotation=20,fontsize=12)
plt.xlabel("Release Year",fontsize=20)
plt.ylabel("Number of Albums",fontsize=20)
plt.title("Topsters Breakdown: Albums per Year",fontsize=20)

plt.bar(x_axis,genrecount["GENRE"],color='teal')

plt.savefig("Visuals/topsters_genrecount.jpg")
plt.show()
```

```python
# Create bar graph showing distribution of albums by country
x_axis=x_axis = np.arange(len(topsters_DF["COUNTRY"].unique()))
tick_loc = [value for value in x_axis]
nations = sorted(topsters_DF["COUNTRY"].unique().tolist())
nationcount = pd.DataFrame(topsters_DF["COUNTRY"].value_counts().sort_index())

plt.figure(figsize=(20,10))
plt.xticks(tick_loc,nations,rotation=20,fontsize=12)
plt.xlabel("Country of Origin",fontsize=20)
plt.ylabel("Number of Albums",fontsize=20)
plt.title("Topsters Breakdown: Albums per Country",fontsize=20)

plt.bar(x_axis,nationcount["COUNTRY"],color='orange')

plt.savefig("Visuals/topsters_countries.jpg")
plt.show()
```

```python
# Create a pie chart breaking down the list by genre
by_genre = topsters_DF.groupby("GENRE")["TITLE"].count()
genre_pie = by_genre.plot(kind="pie",title="Top Albums by Genre")
plt.tight_layout()
```
