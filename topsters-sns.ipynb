```python
# Topsters Practice
import os
import csv
import pandas as pd

topsters_DF = pd.read_csv("Data/topsters.csv")
topsters_DF.head(10)
```

```python
# Import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
# Create a barplot of the number of albums by genre
genre_count = sns.catplot(data=topsters_DF, x='GENRE', kind="count", height=10, aspect=1.75)
genre_count.figure.suptitle(
    "Number of Albums by Genre",
    fontsize=24,
    fontdict={"weight": "bold"}
);
```

```python
year_count = sns.catplot(data=topsters_DF, x='YEAR', kind="count", height=10, aspect=2.2)
year_count.figure.suptitle(
    "Number of Albums by Year",
    fontsize=24,
    fontdict={"weight": "bold"}
);
```
